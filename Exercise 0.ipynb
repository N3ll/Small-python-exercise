# Exercise 0 – Parabola

## Task 1

Write a function that draws a parabola $y_a(x)$ for a given range of values x and parameter a.

\begin{align}
y_a(x) = a \cdot x^2
\end{align}


```python
import matplotlib.pyplot as plt

def draw_parabola_a_x(range_x,a=1):
    x_axis=[]
    y_axis=[]

    for x in range_x:
        y=a*(x**2)
        x_axis.append(x)
        y_axis.append(y)

    fig= plt.figure()
    axes=fig.add_subplot(111)
    axes.plot(x_axis,y_axis)
    plt.show()
    
draw_parabola_a_x(range(-4,5),2)
```

## Task 2

Assume that $a \equiv 1 $. Create a list that holds the $y$ values for $ x \in [-10, …, 10] $.

```python
def y_values_hold(x_range, a=1):
    return [a*(x**2) for x in x_range]

y_values_hold(range(-2,3))
```

## Task 3

Create a dictionary that uses $a \in \{ -2, 1, \frac{1}{2} \} $ as its index and holds the respective $y$ values for $ x \in [-10, …, 10] $.

```python
def y_values_dict(x_range,a_range):
    y=[]
    for a in a_range:
        y.append([a*(x**2) for x in x_range])
    return {k:v for k,v in zip(a_range,y)}


x=range(-10,11)
a=[-2,1,1/2]
y_values_dict(x,a)
```

## Task 4

Use the function to plot 3 parabolas $y_a(x)$ with the parameters $ a \in \{ -2, 1, \frac{1}{2} \} $ over the range $ x \in [-10, 10]$ . Add labels and a legend to your plot!

```python
def draw_parabola(x,y):
    fig = plt.figure()
    axes = fig.add_subplot(111)
    axes.set_xlabel("X values")
    axes.set_ylabel("Y values")
    axes.plot(x,y, label = "parabola")
    plt.legend()
    plt.show()
    
x = range(-10,11)
a = [-2,1,1/2]
dict_y = y_values_dict(x,a)

for y in dict_y.values():
    draw_parabola(x,y)
```
